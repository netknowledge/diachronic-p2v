```python
import pickle

decades = ['1950_to_1959', '1960_to_1969', '1970_to_1979', 
           '1980_to_1989', '1990_to_1999', '2000_to_2009', '2010_to_2021']

walk_lengths = {}
for i, decade in enumerate(decades):
    print(f'Processing decade: {decade}')
    walks = pickle.load(open(f'/media/sdb/p2v/pickles/decades/{decade}/filtered_walks.pkl', 'rb'))
    walk_lengths[decade] = [len(walk) for walk in walks]
```

```python
import seaborn as sns
from matplotlib import pyplot as plt

fig, axes = plt.subplots(figsize=(8, 4), nrows=2, ncols=4, sharex=True, sharey=False)

for i, decade in enumerate(decades):
    ax = axes[i//4, i%4]
    sns.histplot(walk_lengths[decade], binwidth=1, ax=ax)
    ax.set_xlabel('')
    ax.set_ylabel('')
    ax.ticklabel_format(style='sci', axis='y', scilimits=(0,0))

    # hide the last subplot
    for i in range(len(decades), len(axes.flat)):
        axes.flat[i].set_visible(False)

fig.suptitle('Distribution of Walk Lengths by Decade')
fig.supxlabel('Walk Length')
fig.supylabel('Frequency')
plt.tight_layout()
plt.show()
```
